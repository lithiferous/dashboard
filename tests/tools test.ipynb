```python
def format_pct(df, off_x, off_y):
    tdf = df.iloc[off_x:, off_y:].fillna('').replace('#DIV/0! (Function DIVIDE parameter 2 cannot be zero.)', '')
    tdf = tdf.replace(to_replace='\.', value=',', regex=True)
    for idx, row in tdf.iterrows():
        maxlen = len(row)
        new_elements = []
        for cell in range(maxlen):
            print(cond[0] + '%')
            cond = re.search('.(,)..', str(row[cell]))
            if cond != None:
                new_elements.append(cond[0] + '%')
            else:
                new_elements.append(row[cell])
        tdf.loc[idx] = pd.Series(new_elements)
    return tdf
```
