```python
df = pd.read_csv('data/cohort/orders_full.csv', sep = ',', index_col=0)
df = df[['ORDER_DATE', 'TOTAL', 'CLIENT_ID', 'ORDER_ID']]
df.columns = ['date_ord', 'revenue', 'id_cust', 'id_ord']
df['revenue'] = (df['revenue']/1000).astype('int64')
df['date_ord'] = pd.to_datetime(df.date_ord)
df = df.sort_values('date_ord').reset_index(drop=True)
df.to_feather('data/cohort/orders.f')
def append_orders(orders, new_orders):
    new_orders = pd.read_csv(new_orders, sep = ';')
    new_orders = new_orders[['ID –∑–∞–∫–∞–∑–∞', '–î–∞—Ç–∞ —Å–æ–∑–¥–∞–Ω–∏—è', '–°—É–º–º–∞ —Å —É—á–µ—Ç–æ–º –¥–æ—Å—Ç–∞–≤–∫–∏, —Ä—É–±.', 'ID –ö–ª–∏–µ–Ω—Ç–∞']]
    new_orders.columns = ['id_ord', 'date_ord', 'revenue', 'id_cust']
    new_orders['revenue'] = new_orders.revenue.astype('int64')
    new_orders['date_ord'] = pd.to_datetime(new_orders.date_ord)
    new_orders['date_ord'] = new_orders.date_ord.dt.strftime('%Y-%m-%d')
    new_orders['date_ord'] = pd.to_datetime(new_orders.date_ord)
    new_orders = new_orders.sort_values('date_ord').reset_index(drop=True)
    
    orders = pd.read_feather(orders)
    orders = orders.append(new_orders, ignore_index = True)
    orders.to_feather('data/cohort/orders.f')
    return orders
append_orders('data/cohort/orders.f', 'data/cohort/orders_last.csv')
append_orders('data/cohort/orders.f', 'data/cohort/orders_actual.csv')
```
